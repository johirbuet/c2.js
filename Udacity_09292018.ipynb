```python
#importing modules
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#Downloading the file 
df = pd.read_csv('C:/Users/yurin/desktop/noshowappointments-kagglev2-may-2016.csv')
df
```

```python
#overall statiscal data
df.describe()
```

```python
#excluding wierd data
df = df.where(df['Age'] > -1)
df.describe()
```

```python
#changing key name
df_new = df.rename(columns={'No-show':'Noshow'})
df_new
```

```python
#changing Noshow data type to integer
df_new.loc[df_new['Noshow'] == 'Yes', 'Noshow'] = 1
df_new.loc[df_new['Noshow'] == 'No', 'Noshow'] = 0
df_new
```

```python
#changing appointmentday and scheduledday data to datetime data
df_new['AppointmentDay'] = pd.to_datetime(df_new['AppointmentDay'])
df_new['ScheduledDay'] = pd.to_datetime(df_new['ScheduledDay'])
df_new
```

```python
#checkeing datatypes
df_new.dtypes
```

```python
#adding waitingdays column
df_new['Daystoshow'] = df_new['AppointmentDay'] - df_new['ScheduledDay']
df_new
```

```python
#crosstab table of age and noshow data
CT1 = pd.crosstab(df_new['Age'], df_new['Noshow'], margins = True)
CT1
```

```python
#plotting age dependant noshow and show numbers
CT1.plot()
plt.xlim(0,100)
plt.ylim(0,5000)
plt.xticks([0,10,20,30,40,50])
fname ='Entire_age_distribution.png'
plt.savefig(fname)
```

```python
#bar graph of noshow and show rate on gender, scholarship, alcoholism, hipertension, handcap, diabetes, sms_received
#saving figure as png file

CT2 = pd.crosstab(df_new['Gender'], df_new['Noshow'], margins = True, normalize = 'index')
ax = CT2.plot.bar()
ax.set_title('Bar chart')
plt.ylabel('Rate')
plt.ylim(0,1)
fname ='Gender.png'
plt.savefig(fname)

CT3 = pd.crosstab(df_new['Scholarship'], df_new['Noshow'], margins = True, normalize = 'index')
ax = CT3.plot.bar()
ax.set_title('Bar chart')
plt.ylabel('Rate')
plt.ylim(0,1)
fname ='Scholarship.png'
plt.savefig(fname)

CT4 = pd.crosstab(df_new['Alcoholism'], df_new['Noshow'], margins = True, normalize = 'index')
ax = CT4.plot.bar()
ax.set_title('Bar chart')
plt.ylabel('Rate')
plt.ylim(0,1)
fname ='Alcoholism.png'
plt.savefig(fname)

CT5 = pd.crosstab(df_new['Hipertension'], df_new['Noshow'], margins = True, normalize = 'index')
ax = CT5.plot.bar()
ax.set_title('Bar chart')
plt.ylabel('Rate')
plt.ylim(0,1)
fname ='Hipertension.png'
plt.savefig(fname)

CT6 = pd.crosstab(df_new['Handcap'], df_new['Noshow'], margins = True, normalize = 'index')
ax = CT6.plot.bar()
ax.set_title('Bar chart')
plt.ylabel('Rate')
plt.ylim(0,1)
fname ='Handcap.png'
plt.savefig(fname)

CT7 = pd.crosstab(df_new['Diabetes'], df_new['Noshow'], margins = True, normalize = 'index')
ax = CT7.plot.bar()
ax.set_title('Bar chart')
plt.ylabel('Rate')
plt.ylim(0,1)
fname ='Diabetes.png'
plt.savefig(fname)

CT8 = pd.crosstab(df_new['SMS_received'], df_new['Noshow'], margins = True, normalize = 'index')
ax = CT8.plot.bar()
ax.set_title('Bar chart')
plt.ylabel('Rate')
plt.ylim(0,1)
fname ='SMS.png'
plt.savefig(fname)
```
