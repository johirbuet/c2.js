```python
#importing modules
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#downloading file
df = pd.read_csv('C:/Users/yurin/desktop/noshowappointments-kagglev2-may-2016.csv')
df
```

```python
#changing key column name
df_new = df.rename(columns={'No-show':'Noshow'})
df_new
```

```python
#changing noshow data into integer in order to calculate
df_new.loc[df_new['Noshow'] == 'Yes', 'Noshow'] = 1
df_new.loc[df_new['Noshow'] == 'No', 'Noshow'] = 0
df_new
```

```python
#changing appointmentday and scheduledday data into datetime type data
df_new['AppointmentDay'] = pd.to_datetime(df_new['AppointmentDay'])
df_new['ScheduledDay'] = pd.to_datetime(df_new['ScheduledDay'])
df_new
```

```python
#checking datatypes
df_new.dtypes
```

```python
#adding waiting days data to a new column
df_new['Daystoshow'] = df_new['AppointmentDay'] - df_new['ScheduledDay']
df_new
```

```python
#changing sms_received data to boolean in order to extract row data
df_new['SMS_received'] = df_new['SMS_received'].astype(bool)
df_new.dtypes
df_new
```

```python
#trying to extract patients who are receiving SMS
#in addition, extracting patient who didn't and did show up
df_contains_sms_true = df_new.query('SMS_received == True')
df_contains_sms_true_noshow = df_contains_sms_true.query('Noshow == True')
df_contains_sms_true_show = df_new.query('Noshow == False')
```

```python
#mean data of patients who received SMS and showed up
df_contains_sms_true_show.mean()
```

```python
#mean data of patients who received SMS but didn't showed up
df_contains_sms_true_noshow.mean()
```

```python
#trying to extract patients who are not receiving SMS
#in addition, extracting patient who didn't and did show up
df_contains_sms_false = df_new.query('SMS_received == False')
df_contains_sms_false_noshow = df_contains_sms_false.query('Noshow == True')
df_contains_sms_false_show = df_contains_sms_false.query('Noshow == False')
```

```python
#mean data of patients who didn't receive SMS and not showed up
df_contains_sms_false_noshow.mean()
```

```python
#mean data of patients who didn't receive SMS and showed up
df_contains_sms_false_show.mean()
```

```python
#mean data of entire patients who didn't showed up
df_entire_noshow = df_new.query('Noshow == True')
df_entire_noshow.mean()
```

```python
#mean data of entire patients who showed up
df_entire_show = df_new.query('Noshow == False')
df_entire_show.mean()
```
